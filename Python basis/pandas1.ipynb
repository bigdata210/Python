```python
import pandas as pd
url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
titanic = pd.read_csv(url)
titanic.head()
```

```python
titanic["Age"].mean()
```

```python
titanic[["Age","Fare"]].median()
```

```python
titanic[["Age","Fare"]].describe()
```

```python
titanic.agg(
    {
    "Age":["min","max","median","skew"],
    "Fare":["min","max","median","mean"],
     }
)
```

```python
titanic[["Sex","Age"]].groupby("Sex").mean()
```

```python
# 숫자 열을 포함하는 열의 평균을 구하기
titanic.groupby("Sex").mean(numeric_only=True)
```

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
titanic.groupby(["Sex","Pclass"])["Fare"].mean()
```

```python
# value_counts() : 컬럼에서 각 카테고리의 레코드 수를 계산하는 메서드
# 시리즈 객체로 반환됨.
titanic["Pclass"].value_counts()
# 즉, 카테고리 값들과 그 값의 빈도수를 세는 메서드
```

```python
# .count() : "Pclass" 열에서 각 카테고리 값이 몇번 나타났는지 세어서 "Pclass" 값에 따라 그룹화된 결과 반환
# groupby 메서드에 의해 그룹화된 객체 Pclass 열만 선택하여 적용되므로 데이터 프레임으로 반환
titanic.groupby("Pclass")["Pclass"].count()
# 즉. 데이터프레임에서 특정 열을 기준으로 그룹화된 결과
```
